```python
import nltk


# separate patterns for 'validate' case's sentences
def separate_object(split_list):
    
    stopword_list=[",","is","the",'"',"\n"]
    separate_list=[]
    
    d_object=split_list[1]
    temp=split_list[0].split()
    operation=temp.pop(0)
    temp_list=[]
    for w in temp:
        if not w in stopword_list:
            temp_list.append(w)
    info_object=" ".join(temp_list)
    
    separate_list.append(operation)
    separate_list.append(d_object)
    separate_list.append(info_object)
    
    return separate_list



#sorting data as 
#[case,input,validate info,return info,send info,check infor,if info]
def info_sort(values_list):
    count=0
    count_list=[]
    
    #dict_zip[]=values_list
    for i in values_list:
        if i[0]=='@':
            title=i.split('@')[1]
        if i[0]=='#':
            inp=i
            count_list.append(count)
        count+=1


    title_list=[]
    full_list=[]

    title_list.append(title)
    full_list.append(title_list)


    inp_list=[]
    coun1=0
    for j in count_list:
        inp=values_list[j].split("#")[1]
        try:
            new_inp=inp.split('"')[1]
            inp_list.append(new_inp)
        except:
            new_inp=inp
            inp_list.append(new_inp)
                


    info_list=[]
    for i in range(len(count_list)-1):
        info_list.append(values_list[count_list[i]+1:count_list[i+1]])    
    added_info=values_list[count_list[-1]+1:]
    info_list.append(added_info)


    full_list.append(inp_list)

    stopword_list=[",","is","the",'"',"\n"]


    total_info_list=[]

    for i in info_list:
        #print(i)
        temp_info_list=[]
        r_list=[]
        s_list=[]
        v_list=[]
        c_list=[]
        if_list=[]
        for k in i:
            #print(k)

            if k.split()[0] == "Validate":
                #v_list.append(k)
                if len(k.split('of'))==2:
                    info=k.split('of')
                    v_list=separate_object(info)
                if len(k.split('from'))==2:
                    info=k.split('from')
                    v_list=separate_object(info)
                if len(k.split('to'))==2:
                    info=k.split('to')
                    v_list=separate_object(info)
                #v_list.append(sepa_list)
            if k.split()[0] == "Check":
                #c_list.append(k)
                check_all=k.split()
                check_all.pop(0)
                new_check=[]
                for p in check_all:
                    if p not in stopword_list:
                        new_check.append(p)
                check_info=" ".join(new_check)
                c_list.append(check_info)
            if k.split()[0] == "Return":
                r_list.append(k)
            if k.split()[0] == "Send":
                

                temp=k.split()
                s_verb=temp.pop(0)

                merge=" ".join(temp)
                new_info=merge.split('to')

                send_act1=new_info.pop(0)

                if len(new_info[0].split('and'))==2:

                    split_list=new_info[0].split('and')
                    new=split_list[1]
                    new_sp=new.split()
                    send_act2=new_sp.pop(0)


                    s_add=[]
                    for q in new_sp:
                        if q not in stopword_list:
                            s_add.append(q)
                    f_s_add=" ".join(s_add)
                    tot_verb=send_act1+'/'+send_act2
                    tot_info=split_list[0]+'/'+f_s_add
                    s_list.append(tot_verb)
                    s_list.append(tot_info)
                    
                    
                else:
                    send_info1=new_info[0]

                    tot_verb=send_act1
                    tot_info=new_info[0]
                    s_list.append(tot_verb)
                    s_list.append(tot_info)
                    continue
                
                
            if k.split()[0] == "If":

                if_words_list=[]
                words_list=[]

                texts = nltk.word_tokenize(k)
                for w in texts:
                    if not w in stopword_list:
                        words_list.append(w)
                if_words_list.append(words_list[1])
                if_words_list.append(words_list[2])
                if_list.append(if_words_list)

        if len(v_list)==0:
            v_list.append("None")
        if len(c_list)==0:
            c_list.append("None")            
        if len(r_list)==0:
            r_list.append("None")
        if len(s_list)==0:
            s_list.append("None")
        if len(if_list)==0:
            if_list.append("None")

        temp_info_list.append(v_list)
        temp_info_list.append(r_list)
        temp_info_list.append(s_list)
        temp_info_list.append(c_list)
        temp_info_list.append(if_list)

        total_info_list.append(temp_info_list)

    full_list.append(total_info_list)
    
    return full_list


#data to list 
#split each lines
def file_to_list(file_name):
    file=open(file_name,'r')
    hfile=file.read()
    to_list=hfile.split('\n')
    return to_list


#####################################################


info_list=file_to_list('Lab_data_Text.txt')

#Change data as dict.
#And each dict data have each cases data
#[{1:[case1 data],2:[case2 data]...}

dict_zip={}
case_count=0
info_count=0
for i in info_list:
    if i[0]=='@':
        #print(i[1:-1])
        temp_info_count=info_count
        #print(info_count)
        case_count=case_count+1
    info_count=info_count+1
    dict_zip[case_count]=info_list[temp_info_count:info_count]
    

final_list=[]    
for i in dict_zip.values():
    total_list=info_sort(i)
    final_list.append(total_list)
```
